```python
from sentence_transformers import SentenceTransformer
import hdbscan
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Sample stressors (replace with your data)
stressors = [
    "Exams approaching soon",
    "Exams preparation",
    "Exams tomorrow",
    "Deadlines at work",
    "Arguments with family",
    "Job interview stress",
    "Presentation anxiety",
    "Financial concerns",
    "Feeling overwhelmed",
    "Trouble sleeping due to stress",
    "Conflict with friends",
    "Workload pressure"
]

# Step 1: Generate Embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight, good performance
embeddings = model.encode(stressors)

cosine_distances = 1 - cosine_similarity(embeddings)
cosine_distances = cosine_distances.astype('float64') 

# Step 3: Apply HDBSCAN with Precomputed Cosine Distance
clusterer = hdbscan.HDBSCAN(min_cluster_size=2, metric='precomputed')
labels = clusterer.fit_predict(cosine_distances)

# Step 4: Select Anchor Stressors (Centroids)
def select_anchor(cluster_indices):
    cluster_vectors = embeddings[cluster_indices]
    centroid = np.mean(cluster_vectors, axis=0)
    similarities = cosine_similarity([centroid], cluster_vectors)[0]
    anchor_idx = cluster_indices[np.argmax(similarities)]
    return stressors[anchor_idx]


# Step 4: Map Stressors to Anchors
clusters = {}
for label in set(labels):
    if label == -1:
        # Outliers - preserve as unique stressors
        outliers = [stressors[i] for i in range(len(labels)) if labels[i] == -1]
        clusters.update({outlier: outlier for outlier in outliers})
    else:
        indices = [i for i in range(len(labels)) if labels[i] == label]
        anchor = select_anchor(indices)
        for idx in indices:
            clusters[stressors[idx]] = anchor

# Display Results
df = pd.DataFrame(list(clusters.items()), columns=["Original Stressor", "Anchor Stressor"])
print(df)
```
